```python
import random

def rand_result():
  num = random.randint(1,36)
  result = 0
  #green and red
  if(num <= 6):
    result = 1
  #green and blue
  elif(num <= 18):
    result = 10
  #green and green
  elif(num <= 27):
    result = 100
  #blue and red
  elif(num <= 31):
    result = 1000
  #blue and blue
  elif(num <= 35):
    result = 10000
  #red and red
  elif(num <= 36):
    result = 100000
  return result
```

```python
#n: number of boards you create
#k: number of times you roll the dice (size of board)
def roll_dice(n,k):
  frequency = {}
  for i in range(n):
    sum = 0
    for j in range(k):
      sum += rand_result()
    if sum in frequency:
      frequency[sum] += 1
    else:
      frequency[sum] = 1
  print(dict(reversed(sorted(frequency.items(), key=lambda item: item[1]))))

roll_dice(10000000,25)
```
